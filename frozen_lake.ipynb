```python
import gymnasium as gym
from bettermdptools.algorithms.planner import Planner
from bettermdptools.utils.plots import Plots
from gym.envs.toy_text.frozen_lake import generate_random_map
from bettermdptools.utils.test_env import TestEnv
from bettermdptools.algorithms.rl import RL
from bettermdptools.utils.blackjack_wrapper import BlackjackWrapper
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
frozen_lake = gym.make('FrozenLake8x8-v1',desc=generate_random_map(size=4), render_mode=None, is_slippery=True)

# run VI
V_value, V_track_value, pi = Planner(frozen_lake.P).value_iteration(n_iters=2000)

#plot state values
size=(4,4)
Plots.values_heat_map(V_value, "Frozen Lake\nValue Iteration State Values", size)
```

```python
# Clip trailing zeros in case convergence is reached before max iterations
# This is likely when setting the n_iters parameter
max_value_per_iter = np.trim_zeros(np.mean(V_track_value, axis=1), 'b')
Plots.v_iters_plot(max_value_per_iter, "Frozen Lake\nMean Value v Iterations")
```

```python
delta = np.diff(V_track_value, axis=0)
value_differences_abs = np.abs(delta)
mean_value_differences_abs = np.mean(value_differences_abs, axis=1)


# Plotting the mean absolute value function difference vs. iteration
plt.figure(figsize=(10, 6))
plt.plot(mean_value_differences_abs, marker='o', linestyle='-', markersize=4)
# plt.title('Mean Absolute Difference in Value Function vs. Iterations')
plt.xlabel('Iteration')
plt.xlim(0,300)
plt.ylim(-0.01,0.02)
plt.ylabel('Delta')
plt.grid(True)
plt.show()
```
